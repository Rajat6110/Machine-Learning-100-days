```python
num_col = []
cat_col = []
for col in df:
    if  df[col].dtypes == 'object':
        cat_col.append(col)
    elif  df[col].dtypes != 'object':
        num_col.append(col)
```

```python
cat_col
```

```python
num_col
```

```python
plt.figure(figsize = (12,8))
sns.countplot(x = 'ProdTaken' , data = df)
plt.title("Univariate analysis")
plt.show()
```

```python
for col in cat_col:
    
    plt.figure(figsize=(8, 5))
    sns.countplot(x = col, data=df, palette = 'viridis')
    print(f"Univariate Analysis {cat_col}")
    plt.show()
```

```python
for col in num_col:
    plt.figure(figsize= (8,5))
    sns.histplot(x = col , data = df , kde = True , palette = 'viridis')
```
