```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(X),
    columns=X.columns
)
```

```python
X_scaled.drop('id' , axis = 1 , inplace = True)
```

```python
def compute_vif(X_scaled):
    vif = pd.DataFrame()
    vif['feature'] = X_scaled.columns
    vif['VIF'] = [
        variance_inflation_factor(X_scaled.values, i)
        for i in range(X_scaled.shape[1])
    ]
    return vif.sort_values(by='VIF', ascending=False)

vif_df = compute_vif(X_scaled)
print(vif_df)
```

```python
ridge = Ridge(alpha = .001)
```

```python
ridge
```

```python
ridge.fit(X_scaled,y)
```

```python
yyy = ridge.predict(X_scaled)
```

```python
r2_score(y , yyy)
```
